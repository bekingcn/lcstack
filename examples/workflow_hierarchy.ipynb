```python
import lcstack

lcstack.set_config_root("configs")
lcs = lcstack.LcStack.from_yaml("wf_hierarchical_team/graph_supervisor_header.yaml", env=True, secret=True, settings=True)
```

```python
teams_manager = lcs.get("teams_manager")
print(teams_manager)

# from langchain_core.tracers.stdout import ConsoleCallbackHandler
from lcstack.components.tracers.console import ConsoleCallbackHandler
config = {"thread_id": "wf_hierarchical_team_1", "callbacks": [ConsoleCallbackHandler()]} # callbacks=["llm"]

teams_manager.invoke({"question": 
                      ("tell me three jokes first, one about dog and two about apple, but only one at a time;"
                       " and then teach me 2 knowledges, one about dog, and one about apple")},
                       config=config)
```
